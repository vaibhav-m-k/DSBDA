```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset['age'] = dataset['age'].fillna(method='ffill')
display(dataset.isnull().sum())
```

```python
ax = sns.countplot(x='sex', hue='survived', data=dataset)

# Add exact values to each bar
for container in ax.containers:
  ax.bar_label(container)
```

```python
sns.countplot(data=dataset, x="who", hue="survived")
```

```python
ax = sns.countplot(x='pclass', hue='survived', data=dataset)
for container in ax.containers:
  ax.bar_label(container)
```

```python
sns.countplot(data=dataset, x="embarked", hue="survived")
```

```python
sns.boxplot(data=dataset, x="survived", y="fare")
```

```python
sns.boxplot(data=dataset, x="survived", y="age")
```

```python
sns.histplot(data=dataset, x="fare")
```
