```python
import numpy as np
from scipy import stats
import pandas as pd

df = pd.read_csv("/content/StudentsPerformance.csv")

df['math score'] = pd.to_numeric(df['math score'], errors='coerce')

df['math score'].fillna(df['math score'].mean(), inplace=True)

z = np.abs(stats.zscore(df['math score']))
print(z)
```

```python
threshold = 0.18
sample_outliers = np.where(z <threshold)
sample_outliers
```

```python
new_df=df
for i in sample_outliers:
    new_df.drop(i,inplace=True)
new_df
```

```python
sorted_rscore = df['reading score'].dropna().sort_values()
median=np.median(sorted_rscore)
median
refined_df=df


Q1 = refined_df['reading score'].quantile(0.25)
Q3 = refined_df['reading score'].quantile(0.75)
IQR = Q3 - Q1
upr_bound = Q3 + 1.5 * IQR

refined_df['reading score'] = np.where(refined_df['reading score'] > upr_bound, median, refined_df['reading score'])
refined_df
```

```python
lwr_bound = Q1 - 1.5 * IQR
refined_df['reading score'] = np.where(refined_df['reading score'] < lwr_bound, median, refined_df['reading score'])
refined_df
```

```python
col = ['reading score']
refined_df.boxplot(col)
```
