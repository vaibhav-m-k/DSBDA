```python
import pandas as pd
import numpy as np
```

```python
student_df = pd.read_csv("StudentPerformanceFactors.csv")
student_df.head()
```

```python
student_df.info()
student_df.describe()
```

```python
print(student_df.columns)
```

```python
grouped_student = student_df.groupby("Gender")
print("Mean:\n", grouped_student.mean(numeric_only=True))
print("\nMedian:\n", grouped_student.median(numeric_only=True))
print("\nMinimum:\n", grouped_student.min(numeric_only=True))
print("\nMaximum:\n", grouped_student.max(numeric_only=True))
print("\nStandard Deviation:\n", grouped_student.std(numeric_only=True))
```

```python
male_attendance = student_df[student_df["Gender"]=="Male"]["Attendance"].tolist()
female_attendance = student_df[student_df["Gender"]=="Female"]["Attendance"].tolist()
print("Male Age List:", male_attendance)
print("Female Age List:", female_attendance)
```
