```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("Boston.csv")
df
```

```python
df.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
df.drop(['a'],axis=1, inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.distplot(df['medv'], bins=20)
plt.show()
```

```python
df.corr().round(2)
```

```python
plt.figure(figsize=(15, 10))
plt.title("rm")
plt.xlabel("rm")
plt.ylabel('medv')
plt.scatter(df['rm'],df['medv'])
```

```python
plt.figure(figsize=(15, 10))
plt.title("lstat")
plt.xlabel("lstat")
plt.ylabel('medv')
plt.scatter(df['lstat'],df['medv'])
```

```python
rm= df['rm']
medv= df['medv']
x_train, x_test, y_train, y_test = train_test_split(rm, medv, test_size = 0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
x_train = np.array(x_train).reshape(-1, 1)
y_train = np.array(y_train).reshape(-1, 1)
x_test = np.array(x_test).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
plt.title("predicted results")
plt.xlabel("rm")
plt.ylabel('medv')
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_predict, color='red',label="aa")
plt.scatter(x_test, y_test, color="yellow")
```

```python
lstat= df['lstat']
x_train, x_test, y_train, y_test = train_test_split(lstat, medv, test_size = 0.2)
x_train = np.array(x_train).reshape(-1, 1)
y_train = np.array(y_train).reshape(-1, 1)
x_test = np.array(x_test).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
plt.title("predicted results")
plt.xlabel("lstat")
plt.ylabel('medv')
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_test, color="yellow")
```
